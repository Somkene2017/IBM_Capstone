## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT *
FROM DatabaseDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df2 = pd.read_sql_query(QUERY,conn)

df_top_database = df2['DatabaseDesireNextYear'].value_counts()[0:5].to_frame()
df_top_database
```

```python
# your code goes here
QUERY = """
SELECT *
FROM LanguageDesireNextYear
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df3 = pd.read_sql_query(QUERY,conn)

df_top_language = df3['LanguageDesireNextYear'].value_counts()[:5].to_frame()
df_top_language
```

```python
# your code goes here
QUERY = """
SELECT *
FROM DatabaseWorkedWith
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df3 = pd.read_sql_query(QUERY,conn)

df_top_language = df3['DatabaseWorkedWith'].value_counts().to_frame()
df_top_language
```

```python
QUERY = """
SELECT *
FROM DevType
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df3 = pd.read_sql_query(QUERY,conn)

df_top_language = df3['DevType'].value_counts().to_frame()
df_top_language
```

```python

df_top_database.plot(kind='pie', y='DatabaseDesireNextYear',
                            autopct='%1.1f%%', # add in percentages
                            startangle=90,     # start angle 90° (Africa)
                            shadow=True,       # add shadow      
                            )

plt.title('Top 5 Database')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.legend(labels=df_top_database.index, loc='upper left') 


plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

# Filter for age group 30 to 35
df_filtered = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate median WorkWeekHrs and CodeRevHrs
df_filtered_group = df_filtered.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median().reset_index()
median_values = df_filtered[['WorkWeekHrs', 'CodeRevHrs']].median()
# Create the stacked chart
median_values.plot(kind='bar', stacked=True)
df_filtered_group.plot(x='Age', kind='bar', stacked=True)
# Add labels and title
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age Group')
plt.ylabel('Hours')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df_filtered = df[(df['Age'] >= 25) & (df['Age'] <= 30)]

# Calculate median WorkWeekHrs and CodeRevHrs
df_filtered_group = df_filtered.groupby('Age')['ConvertedComp'].median()

# Create the Line chart
df_filtered_group.plot( kind='line')

# Add labels and title
plt.title('Median ConvertedComp for Age Group 25-30')
plt.xlabel('Age Group')
plt.ylabel('ConvertedComp')

plt.show()
```
